```python
# instale  a biblioteca  antes de rodar:
# pip install deep-translator

!pip install deep-translator

from deep_translator import  GoogleTranslator

# Texto original
texto = "Hello, world ! How are you today?"

# Dicionario com os idiomas desejados (nome -> código ISO)
idiomas = {
    "Português": "pt",
    "Russo": "ru",
    "Coreano": "ko",
    "Francês": "fr",
    "Alemão": "de",
    "Inglês": "en",
    "Espanhol (México)": "es", # o google usa 'es' para espanhol geral
    "Italiano": "it"

}

print("text original:", texto)
print("-" * 50)

#traduz para cada idioma da lista
for idioma, codigo in idiomas.items():
    traducao = GoogleTranslator(source='auto', target=codigo).translate(texto)
    print(f"{idioma}: {traducao}")



```
